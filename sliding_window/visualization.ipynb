```python
import pandas as pd
import numpy as np

```

```python
import pickle
def makefile(what,filename):
    with open(filename,"wb") as f3:
        pickle.dump(what,f3)

def readfile(filename):
    with open(filename,"rb") as f4:
        ans=pickle.load(f4)
    return ans
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
change=[60,120,180,240,300,420,480,540,600,660,720]
```

```python
v1=[1,3,6,7,8,9,11,13,17]
```

```python
for i in range(20):
    if i+1 in v1:
        error=readfile("E_test660_c"+str(i+1)+".pkl")
        d=1000
        #error_min=np.array([error[n:n+d].min() for n in range(len(error)-d) ])
        plt.figure()
        for v in range(len(change)):
            plt.vlines(change[v]*1000, error.min(),error.max(),color="red",linestyle = "dashed")
        plt.plot(error)
```

```python
for i in range(20):
    if i+1 in v1:
        error=readfile("E_test660_c"+str(i+1)+".pkl")
        d=1000
        error_min=np.array([error[n:n+d].min() for n in range(len(error)-d) ])
        plt.figure()
        plt.title("Local minimum construction error")
        for v in range(len(change)):
            plt.vlines(change[v]*1000, error_min.min(),error_min.max(),color="red",linestyle = "dashed")
        plt.plot(error_min)
```

```python
index=np.array([n for n in range(len(error_min))])/1000
```
